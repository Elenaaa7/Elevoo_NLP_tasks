```python
import pandas as pd

resume_df = pd.read_csv("/content/unzipped_folders/resume/Resume/Resume.csv")
resume_df.head()
```

```python
job_df = pd.read_csv("/content/unzipped_folders/job/job_descriptions.csv")
job_df.head()
```

```python

# Keep only the clean text
resume_df = resume_df[["ID", "Resume_str", "Category"]].dropna()
resume_texts = resume_df["Resume_str"].tolist()
```

```python
# Select only useful columns
job_df = job_df[["Job Id", "Job Title", "Role", "Job Description", "skills"]]

# Sample for testing (say 500 jobs instead of 1.6M)
job_df = job_df.sample(500, random_state=42).reset_index(drop=True)

# Combine text into one field for embeddings
job_df["job_text"] = (
    job_df["Job Title"].fillna('') + " " +
    job_df["Role"].fillna('') + " " +
    job_df["Job Description"].fillna('') + " " +
    job_df["skills"].fillna('')
)

job_texts = job_df["job_text"].tolist()
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("all-MiniLM-L6-v2")

# Encode resumes and jobs
resume_embs = model.encode(resume_texts, convert_to_tensor=True, normalize_embeddings=True)
job_embs = model.encode(job_texts, convert_to_tensor=True, normalize_embeddings=True)

# Example: compare the first job to all resumes
import torch

cos_sim = util.cos_sim(job_embs[0], resume_embs)  # similarity for job 0
top_scores, top_idx = torch.topk(cos_sim, k=5)   # get top 5 resumes

print("Job:", job_df.iloc[0]["Job Title"])
for score, idx in zip(top_scores[0], top_idx[0]):
    idx = int(idx)  # ✅ convert tensor to Python int
    print(f"Resume {resume_df.iloc[idx]['ID']} | Category: {resume_df.iloc[idx]['Category']} | Score: {score:.2f}")
```

```python
import re

SEED_SKILLS = {
    "python","pandas","numpy","sql","scikit-learn","tensorflow","pytorch","nlp",
    "power bi","tableau","excel","spark","hadoop","git","docker","aws","azure",
    "ml","deep learning","computer vision","statistics","linux","matplotlib"
}

def normalize_token(t: str) -> str:
    return re.sub(r'[^a-z0-9\+\#\.\- ]', '', t.lower()).strip()

def extract_skills(text: str):
    t = normalize_token(text)
    found = set()
    for s in SEED_SKILLS:
        if s in t:
            found.add(s)
    return sorted(found)
skills = extract_skills(resume_df.iloc[0]["Resume_str"])
print(skills)
```

```python
idx = top_idx[0][0].item()  # يحول tensor → int
resume_skills = set(extract_skills(resume_df.iloc[idx]["Resume_str"]))
```

```python
job_skills = set(extract_skills(job_df.iloc[0]["skills"]))

print("Job skills:", job_skills)
print("Resume matched skills:", job_skills.intersection(resume_skills))
```

```python
results = []
for j in range(len(job_df)):
    sims = util.cos_sim(job_embs[j], resume_embs).cpu().numpy().ravel()
    top5_idx = sims.argsort()[-5:][::-1]
    for i in top5_idx:
        results.append({
            "Job Id": job_df.iloc[j]["Job Id"],
            "Job Title": job_df.iloc[j]["Job Title"],
            "Resume ID": resume_df.iloc[i]["ID"],
            "Resume Category": resume_df.iloc[i]["Category"],
            "Similarity": sims[i]
        })

results_df = pd.DataFrame(results)
results_df.head(10)
```

```python
import streamlit as st
import docx2txt, fitz, os
from sentence_transformers import SentenceTransformer, util
import pandas as pd

# Load preprocessed jobs
job_df = pd.read_csv("/content/unzipped_folders/job/job_descriptions.csv")
job_df["job_text"] = (
    job_df["Job Title"].fillna('') + " " +
    job_df["Role"].fillna('') + " " +
    job_df["Job Description"].fillna('') + " " +
    job_df["skills"].fillna('')
)

model = SentenceTransformer("all-MiniLM-L6-v2")
job_embs = model.encode(job_df["job_text"].tolist(), convert_to_tensor=True, normalize_embeddings=True)

def read_resume(uploaded_file):
    text = ""
    if uploaded_file.name.endswith(".txt"):
        text = uploaded_file.read().decode("utf-8")
    elif uploaded_file.name.endswith(".pdf"):
        with fitz.open(stream=uploaded_file.read(), filetype="pdf") as doc:
            for page in doc:
                text += page.get_text()
    elif uploaded_file.name.endswith(".docx"):
        temp_path = "temp_resume.docx"
        with open(temp_path, "wb") as f:
            f.write(uploaded_file.read())
        text = docx2txt.process(temp_path)
        os.remove(temp_path)
    return text

def match_resume(resume_text, top_n=5):
    resume_emb = model.encode([resume_text], convert_to_tensor=True, normalize_embeddings=True)
    cos_sim = util.cos_sim(resume_emb, job_embs)[0]
    top_scores, top_idx = cos_sim.topk(top_n)

    results = []
    for score, idx in zip(top_scores, top_idx):
        job = job_df.iloc[int(idx)]
        results.append({
            "Job Title": job["Job Title"],
            "Role": job["Role"],
            "Similarity Score": round(float(score) * 100, 2)
        })
    return pd.DataFrame(results)

# --- Streamlit UI ---
st.title("📄 Resume Screening with NLP")

uploaded_file = st.file_uploader("Upload your resume", type=["txt", "pdf", "docx"])

if uploaded_file:
    resume_text = read_resume(uploaded_file)
    st.subheader("Resume Preview")
    st.write(resume_text[:1000] + "...")  # show first 1000 chars

    st.subheader("Top Job Matches")
    results_df = match_resume(resume_text, top_n=5)
    st.dataframe(results_df)
```
